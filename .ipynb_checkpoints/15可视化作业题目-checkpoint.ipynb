```python
import seaborn as sns
import numpy as np

#load dataset
car_crashes=sns.load_dataset('car_crashes')
```

```python
#car_crashes head
print(car_crashes.head())
```

```python
sns.pairplot(data=car_crashes)
```

```python
#bivariables plot
sns.jointplot(data=car_crashes,x='total',y='alcohol',kind='scatter')
sns.jointplot(data=car_crashes,x='total',y='alcohol',kind='kde')
sns.jointplot(data=car_crashes,x='total',y='alcohol',kind='hex')
```
