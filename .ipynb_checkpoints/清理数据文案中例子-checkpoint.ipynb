```python
import pandas as pd
df=pd.read_csv('/Users/bingqihou/Downloads/accountMessage.csv',index_col=0)
df
```

```python
df.rename(columns={'0':'Number','1':'Name','2':'Age','3':'Weight','4':'M1','5':'M2','6':'M3'
                  ,'7':'F1','8':'F2','9':'F3'},inplace=True)
```

```python
rows_with_lbs=df['Weight'].str.contains('lbs').fillna(False)
df[rows_with_lbs]
for i,lbs_row in df[rows_with_lbs].iterrows():
    # 截取从头开始到倒数第三个字符之前，即去掉lbs
    weight=int(float(lbs_row['Weight'][:-3])/2.2)
    df.at[i,'Weight']='{}kgs'.format(weight)
```

```python

df.dropna(how='all',inplace=True)
df
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Age']=round(df['Age'],0)
```

```python
df['Weight'].fillna(df['Weight'].value_counts().index[0],inplace=True)
```

```python
df
```

```python
df[['First_name','Last_name']]=df['Name'].str.split(expand=True)
df.drop(['Name'],axis=1,inplace=True)
df
```

```python
df.drop_duplicates(['First_name','Last_name'],inplace=True)
```

```python
df
```

```python
# 移动first_name和last_name这俩列
first_name=df.pop('First_name')
df.insert(1,'First_name',first_name)
last_name=df.pop('Last_name')
df.insert(2,'Last_name',last_name)
```
